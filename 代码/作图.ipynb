```
def fun1(str):
  str1 = str.split('1110')
  str2 = str1[0]
  for i in range(len(str1) - 1):
    str2 = str2 + '111' + str1[i+1]
  return str2

fun1('0111010111101')
```

```
def fun2(str):
  str1 = str.split(' ')
  num = 0
  for i in range(len(str1)):
    for j in range(len(str1[i]) - 1):
      if str1[i][j] == str1[i][j+1]:
        num = num + 1
        break
  return num

fun2('apple banana cherry orange')
```
